# Import necessary libraries



```python
import pandas as pd
# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('iris_csv.csv')
```

Analyze dataset to find information about the instances


```python
dataframe.head()
```

```python
# to check for any NULL values
dataframe.isnull().sum()
```

```python
# to check for the number of class in the dataset
number_of_class = dataframe.drop_duplicates(subset='class',)
number_of_class
```

```python
# to check total count of each class present
dataframe.value_counts("class")
```

```python
# groupby to check if balanced or imbalanced
print(dataframe.groupby('class').size().reset_index())
```

```python
# Visualize dataset according to class label and total instances
sns.countplot(x='class',data=dataframe)
plt.show()
```

```python
# to check relationship between sepallength and sepalwidth 
sns.scatterplot(x='sepallength', y='sepalwidth', hue='class', data=dataframe, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```


Above plot shows that: 
*   Setosa has smaller sepal lengths but larger sepal widths.
*   Versicolor lies in the middle of the other two species in terms of sepal length and width
*   Virginica has larger sepal lengths but smaller sepal widths.





```python
# to check relationship between petallength and petalwidth 
sns.scatterplot(x='petallength', y='petalwidth',hue='class', data=dataframe, )
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```

Above plot confirms that:


*   Setosa has the smallest petallength and petalwidth among the three classes
*   Virginica has the largest petallength and petalwidth among the three classes
*   Versicolor has average petallength and petalwidth among the three classes


```python
# pairplots are good way for multivariate analysis (more than two dependent variables)
sns.pairplot(dataframe, hue='class', height=2)
```

Histograms allow seeing the distribution of data for various columns. It can be used for uni as well as bi-variate analysis

```python
# outline for the four histograms using subplots
fig, axes = plt.subplots(2, 2, figsize=(10,10))
 
axes[0,0].set_title("Sepal Length")
axes[0,0].hist(dataframe['sepallength'])
 
axes[0,1].set_title("Sepal Width")
axes[0,1].hist(dataframe['sepalwidth']);
 
axes[1,0].set_title("Petal Length")
axes[1,0].hist(dataframe['petallength']);
 
axes[1,1].set_title("Petal Width")
axes[1,1].hist(dataframe['petalwidth']);
```

From the above plot, we can see that – 


*   The highest frequency of the sepal length is between 30 and 35 which is between 5.5 and 6
*   The highest frequency of the sepal Width is around 70 which is between 3.0 and 3.5
*   The highest frequency of the petal length is around 50 which is between 1 and 2
*   The highest frequency of the petal width is between 40 and 50 which is between 0.0 and 0.5

```python
# to check for pairwise correlation in the dataset
dataframe.corr(method='pearson')
```
